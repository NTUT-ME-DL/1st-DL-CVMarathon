利用實際的影像，嘗試自己搭建一個 1 × 1 和 3 × 3 的模型    
看通過 1 × 1 和 3 × 3 卷積層後會有什麼變化？
大家可以自己嘗試著搭建不同層數後，觀察圖形特徵的變化

```python
import cv2
import matplotlib.patches as patches
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from keras import backend as K
from keras.models import Sequential 
from keras.layers.core import Activation
from keras.layers.convolutional import Conv2D, MaxPooling2D
```

```python
image = cv2.imread('image/dog.jpg')

def show(img):
    # plt.imshow 預設圖片是 rgb 的
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()

show(image)
```

```python
model = Sequential()
model.add(Conv2D(3, (1, 1), padding="same", data_format='channels_last', activation='relu', input_shape=image.shape))

# 接續搭建一個 4 個 3 * 3 的 filters 
model.add(Conv2D(4, (3, 3), padding="same", data_format='channels_last', activation='relu', input_shape=image.shape))

print(model.summary())
```

```python
image_batch = np.expand_dims(image, axis=0)
print(image_batch.shape)
```

```python
image_conv = model.predict(image_batch)
img = np.squeeze(image_conv,axis=0)

print(img.shape)
plt.imshow(img)
```
