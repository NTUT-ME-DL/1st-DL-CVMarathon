YOLO 損失函數

* 損失函數是描述模型預測出來的結果和實際的差異的依據
* YOLO 損失函數的設計包含物件位置的定位和類別辨識
* YOLO 損失函數透過超參數來使模型有不同的辨識能力

Bounding box 的寬高和中心的計算損失方式有什麼不一樣？為什麼要有不同的設計？

中心計算損失：使用兩個點距離的差異量

寬高計算損失：使用兩個框長寬的差異量，但是長寬有開根號

需要注意的是：用寬和高的開根號代替原來的寬和高，是因爲相同寬和高的誤差對於 小的目標 精度影響比 大的目標 要大

舉個例子，原來 w = 10，h = 20，預測出來 w = 8，h = 22，跟原來 w = 3，h = 5，預測出來 w = 1，h = 7 相比，其實前者的誤差要比後者小，
但是如果不加開根號，那麼損失都是一樣：4 + 4 = 8，但是加上根號後，變成 0.15 和 0.7。
